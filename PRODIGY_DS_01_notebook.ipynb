```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Skip the first 4 metadata rows
df = pd.read_csv("C:\\Users\\ASUS\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv",skiprows=4)

df.head()
```

```python
# Select the latest population

# Change the year if required.
year = "2022"
```

```python
# Keep only Country Name and selected year
population = df[["Country Name", year]]
```

```python
# Remove rows with missing population values
population = population.dropna()
```

```python
# Sort countries by population
population = population.sort_values(by=year, ascending=False)
```

```python
# Select top 10 countries
top10 = population.head(10)
```

# Create Bar Chart

```python
plt.figure(figsize=(12, 6))

plt.bar(top10["Country Name"],
        top10[year])

plt.title("Top 10 Most Populated Countries (2022)", fontsize=15)
plt.xlabel("Country")
plt.ylabel("Population")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
